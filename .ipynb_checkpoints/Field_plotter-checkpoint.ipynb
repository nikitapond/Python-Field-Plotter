# Field plotter
Take any vector or scalar field of up to 3 dimensions (x,y,z)

There are multiple steps to do this:

1) Input of equation
The first code segment only takes a single equation, I.E a scaler field, with x,y, and z variables.
To go from a string to an equation, the string is itterated through. Each seperate term is then found.

```python
import numpy as np

class Term:  
    
    class Misc:
        e = np.exp(1)
        pi = np.pi
    
    ERROR=-1
    X=0
    Y=1
    Z=2
    CONST=3
    ADD=4
    SUBTRACT=5
    MULT=6
    DIVIDE=7
    POW=8
    
    def __init__(self, a, b, Func, c1=0, c2=0):
        self.a=a
        self.b=b
        self.f=Func   
        self.c1=c1
        self.c2=c2
        
    
    def Value(self,x,y,z):
        f=self.c1
        s=self.c2        
        if(type(self.a)==Term):
            f=self.a.Value(x,y,z)
        elif(self.a==Term.X):
            f=x
        elif(self.a==Term.Y):
            f=y
        elif(self.a==Term.Z):
            f=z
    
        if(type(self.b)==Term):
            s=self.b.Value(x,y,z)
        if(self.b==Term.X):
            s=x
        elif(self.b==Term.Y):
            s=y
        elif(self.b==Term.Z):
            s=z
        if(self.f==Term.ADD):            
            return f+s
                
        elif(self.f==Term.SUBTRACT):
            return f-s
                
        elif(self.f==Term.MULT):
            return f*s
                
        elif(self.f==Term.DIVIDE):
            return f/s
                
        elif(self.f==Term.POW):
            return f**s
        
        return ERROR
    
    def toString(self):
        string = ""
        if(type(self.a)==Term):
            string = "("+self.a.toString()+")"
        elif(self.a==Term.CONST):
            string = str(self.c1)
        else:
            string=FuncToChar(self.a)
            
        string += FuncToChar(self.f)
        
        if(type(self.b)==Term):
            string += "("+self.b.toString()+")"
        elif(self.b==Term.CONST):
            string += str(self.c2)
        else:
            string+=FuncToChar(self.b)
              
        return string
    
    def __str__(self):
        return self.toString()
def FuncToChar(f):
    if(f==Term.ADD):
        return '+'
    if(f==Term.SUBTRACT):
        return '-'
    if(f==Term.MULT):
        return '*'
    if(f==Term.DIVIDE):
        return'/'
    if(f==Term.POW):
        return '^'
    if(f==Term.X):
        return 'x'
    if(f==Term.Y):
        return 'y'
    if(f==Term.Z):
        return 'z'
    return "?"
def FindFunc(char):
    if(char=='+'):
        return Term.ADD
    elif(char=='-'):
        return Term.Subtract
    elif(char=='*'):
        return Term.MULT
    elif(char=='/'):
        return Term.DIVIDE
    elif(char=='^'):
        return Term.POW
    elif(char.isdigit()):
        return Term.CONST
    elif(char=='x'):
        return Term.X
    elif(char=='y'):
        return Term.Y
    elif(char=='z'):
        return Term.Z  
    
    return Term.ERROR

class Equation:
    
    def SubArray(self, charEq, b, e):
        out = []
        for i in range(b,e+1):
            out.append(charEq[i])
        return out
    
    def SingleTerm(self, charEq,TermIndexi):
        if(FindFunc(charEq[TermIndexi[0]+2])!=Term.ERROR):
            a = FindFunc(charEq[TermIndexi[0]+1])
            b = FindFunc(charEq[TermIndexi[1]+3])
            c1_ = c2_ = 0
            if(a==Term.CONST):
                c1_ = int(charEq[TermIndexi[0]+1])
            if(b==Term.CONST):
                c2_ = int(charEq[TermIndexi[1]+3])
            #print("charEQ:", "".join(charEq), c1_, c2_)
            return Term(a, b, FindFunc(charEq[1]), c1=c1_,c2=c2_)
        else:
            print("An error has occured, middle of ", "".join(charEq), " should be a function")
    
    def TermLength(self, TermIndexi):
        return TermIndexi[1]-TermIndexi[0]+1
    
    def FindClose(self, charEq, startI):
        
        t = 0
        for i in range(startI, len(charEq)):
            if(charEq[i]=="("):
                t+=1
            elif(charEq[i]==")"):
                t-=1
            if(t==0):
                return i
        return -1
    
    #def CombineTerms(self, charE)
    
    def SearchTermIndex(self, i):
        
        for c in range(0,len(self.TermIndex)):
            if(self.TermIndex[i][0] <= i <= self.TermIndex[i][1]):
                return c
        return -1
    
    def CreateTerms(self, charEq, b=0, end=-1):
        if(end==-1):
            end=len(charEq)
        print("CharEq: ", charEq)
        #Step 1, seperate into all individual terms:
        #Step 1a, first find all things in brackets
        
        i=b
        while i < end:
            if(charEq[i]=="("):
                close = self.FindClose(charEq, i)
                Ti = [i,close]
                self.TermIndex.append(Ti)
                i=close
            i+=1
        print(self.TermIndex)

    def __init__(self, equation):
        CharEq = []
        for c in equation:
            if(c.isspace()==False):
                CharEq.append(c)
        self.Terms = []
        self.TermIndex = []
        self.Equation=self.CreateTerms(CharEq)
        #print(self.Equation.toString())
    
    def Value(self, x,y,z):
        return self.Equation.Value(x,y,z)
    
equation = Equation("(2*x)^((y*z)+1)*(x+4)")

#print(equation.Value(3,5,6))
```

Bellow, we plot a 2D scalar field.

```python
import numpy as np
from vpython import quad, vertex, scene, vector, cross, canvas
def fxy(x,y):
    #f(x,y) = e^(x^2 + y^2)
    return np.exp(x**2 + y**2)

xMin=-1
xMax=1
yMin=-1
yMax=1
resolution = 0.1
xS = int((xMax-xMin)/resolution)+1
yS = int((yMax-yMin)/resolution)+1
data = [[None for x in range(xS)] for y in range(yS)]
print(data[0][0])
for i in range(0, xS):
    x=i*resolution + xMin
    for ii in range(0, yS):
        y=ii*resolution + yMin
        
        data[i][ii] = vector(x,y,fxy(x,y))


scene = canvas(width=640, height=480, center=vector(0,0,0))
for i in range(0, xS-1):
    for ii in range(0, yS-1):
        N = cross(data[i+1][ii]-data[i][ii],data[i][ii+1]-data[i][ii]).norm()
        A=vertex(pos=data[i][ii], normal=N)
        B=vertex(pos=data[i+1][ii], normal=N)
        C=vertex(pos=data[i+1][ii+1], normal=N)
        D=vertex(pos=data[i][ii+1], normal=N)
        Q=quad(vs=[A,B,C,D])
```

Below we plot a 3D scalar field. To do this, a value is calculated for the maximum and minimum value for the given range, and a colour is atrributed to each vertex based on the value at this point.

```python
import numpy as np
from vpython import *
from PIL import Image

def phi_xyz(x,y,z):
    #f(x,y) = e^-(x^2 + y^2)
    return np.exp(-(x**2 + y**2 + z**2))

xMin=-1
xMax=1
yMin=-1
yMax=1
zMin=-1
zMax=1
resolution = 0.2
xS = int((xMax-xMin)/resolution)+1
yS = int((yMax-yMin)/resolution)+1
zS = int((zMax-zMin)/resolution)+1
data = [[[None for x in range(xS)] for y in range(yS)] for z in range(zS)]
needMinMax = True
minValue=0
maxValue=0
for i in range(0, xS):
    x=i*resolution + xMin
    for ii in range(0, yS):
        y=ii*resolution + yMin
        for iii in range(0,zS):
            z=iii*resolution + zMin                
            data[i][ii][iii] = phi_xyz(x,y,z)
            if(needMinMax):
                    minValue = data[i][ii][iii]
                    maxValye = minValue
                    needMinMax=False
            elif(data[i][ii][iii]<minValue):
                minValue = data[i][ii][iii]
            elif(data[i][ii][iii]>maxValue):
                maxValue = data[i][ii][iii]
                

scene = canvas(width=640, height=480, center=vector(0,0,0))
scene.lights = []
scene.ambient=vector(0.6,0.6,0.6)
#scene.userzoom =False
scene.autocenter =True
im = Image.open('grad_colour.png') # Can be many different formats.
pix = im.load()           
c=pix[5,0]
def colourGrad(value):
    s = 511-int(508*(value-minValue)/(maxValue-minValue))
    try:
        c=pix[s,0]
    except:
        print(s)
        return vector(0,0,0)
    return vector(c[0],c[1],c[2])/255

for i in range(0, xS):
    x=i*resolution + xMin
    for ii in range(0, yS):
        y=ii*resolution + yMin
        for iii in range(0,zS):
            z=iii*resolution + zMin                
            #S=simple_sphere(pos=vector(x,y,z), radius=resolution*0.25, color=colourGrad(data[i][ii][iii]))
            s = resolution*0.25
            c=box(pos=vector(x,y,z), size=vector(resolution*0.25,resolution*0.25,resolution*0.25), color=colourGrad(data[i][ii][iii]))

Lt = label(pixel_pos=True, align='left',pos=vector(20, 20, 0), text=("Min Value: "+str("%.2f" % minValue)))
Ut = label(pixel_pos=True, align='left',pos=vector(20, scene.height-40, 0), text=("Max Value: "+str("%.2f" % maxValue)))
for i in range(0,52):
    s= 511-int(508*(i/52))
    p=pix[s,0]
    c=vector(p[0],p[1],p[2])/255
    L = label(pixel_pos=True, pos=vector(20,40+i*(scene.height-80)/52,0),color=c, text="", border=False)
shouldRun = True
b=box(pos=vector(0,0,0), texture="grad_color.png")
omega = np.pi/30
t=0
CurrentPos = label(pixel_pos=True, align='right', pos=vector(scene.width-20, 20, 0), text="Camera pos: ")
CurrentFr = label(pixel_pos=True, align='right', pos=vector(scene.width-20, 50, 0), text="f(r) at camera pos: ") 
while(shouldRun):
    rate(100)
    t+=0.01
    scene.forward = vector(np.sin(omega*t), 0, -np.cos(omega*t))
    cPos = scene.camera.pos
    CurrentPos.text = ("Camera pos: <" + str("%.2f" % cPos.x)+","+str("%.2f" % cPos.y)+","+str("%.2f" % cPos.z)+">")
    CurrentFr.text = ("f(r) at camera pos:" + str("%.2f" % phi_xyz(scene.camera.pos.x,scene.camera.pos.y,scene.camera.pos.z)))


    
```

```python
import numpy as np
from vpython import *
from PIL import Image

def fxyz(x,y,z):
    return vector(-y*x,y,0)
xMin=-1
xMax=1
yMin=-1
yMax=1
zMin=-1
zMax=1
resolution = 0.2
xS = int((xMax-xMin)/resolution)+1
yS = int((yMax-yMin)/resolution)+1
zS = int((zMax-zMin)/resolution)+1
data = [[[None for x in range(xS)] for y in range(yS)] for z in range(zS)]
magn =  [[[0 for x in range(xS)] for y in range(yS)] for z in range(zS)]
needMinMax = True
minValue=0
maxValue=0

for i in range(0, xS):
    x=i*resolution + xMin
    for ii in range(0, yS):
        y=ii*resolution + yMin
        for iii in range(0,zS):
            z=iii*resolution + zMin                
            data[i][ii][iii] = fxyz(x,y,z)
            magn[i][ii][iii]=mag(data[i][ii][iii])
            #print(x,y,z, ": ", magn[i][ii][iii])
            if(needMinMax):
                    minValue = magn[i][ii][iii]
                    maxValye = minValue
                    needMinMax=False
            elif(magn[i][ii][iii]<minValue):
                minValue = magn[i][ii][iii]
            elif(magn[i][ii][iii]>maxValue):
                maxValue = magn[i][ii][iii]

scene = canvas(width=640, height=480, center=vector(0,0,0))
scene.lights = []
scene.ambient=vector(0.6,0.6,0.6)
scene.autocenter =True

im = Image.open('grad_colour.png') # Can be many different formats.
pix = im.load()           
c=pix[5,0]
print(minValue, maxValue)
def colourGrad(value):
    s = 0
    c = 0
    try:
        s = 511-int(508*(value-minValue)/(maxValue-minValue))
        c=pix[s,0]
    except:
        return vector(0,0,0)
    return vector(c[0],c[1],c[2])/255

for i in range(0, xS):
    x=i*resolution + xMin
    for ii in range(0, yS):
        y=ii*resolution + yMin
        for iii in range(0,zS):
            z=iii*resolution + zMin                
            a=arrow(pos=vector(x,y,z), size=vector(resolution,resolution,resolution), axis=data[i][ii][iii], color = colourGrad(magn[i][ii][iii]))

            
Lt = label(pixel_pos=True, align='left',pos=vector(20, 20, 0), text=("Min Value: "+str("%.2f" % minValue)))
Ut = label(pixel_pos=True, align='left',pos=vector(20, scene.height-40, 0), text=("Max Value: "+str("%.2f" % maxValue)))
for i in range(0,52):
    s= 511-int(508*(i/52))
    p=pix[s,0]
    c=vector(p[0],p[1],p[2])/255
    L = label(pixel_pos=True, pos=vector(20,40+i*(scene.height-80)/52,0),color=c, text="", border=False)

    
running=True
while(running):
    rate(100)
    p=scene.camera.pos 
    if(p.x >= xMin and p.x<=xMax and p.y>=yMin and p.y<=yMax and p.z>=zMin and p.z<=zMax):
        scene.camera.pos = p+fxyz(p.x,p.y,p.z)
        print(p , str(fxyz(p.x,p.y,p.z)))
        
```
